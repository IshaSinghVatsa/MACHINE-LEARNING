```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.impute import SimpleImputer
df = pd.read_csv("/content/forestfires1.csv",header=0)

#DISPLAYING ORIGINAL DATAFRAME
display(df)
```

**A)** Here Bar chart is plotted to show relation between two numerical columns Rain and Temoeratue and it can be seen that when Temperatute is highest then humidity increases so Rainfall is highest and when temperature is lowest then atmosphere is dry as result rainfall is very very low

```python
import matplotlib.pyplot as plt
  
X = list(df.iloc[:, 8])
Y = list(df.iloc[:, 11])
  
# Plot the data using bar() method
plt.bar(X, Y, color='g')
plt.title("Temp depending on Rain ")
plt.xlabel("Temperature")
plt.ylabel("Rain")
  
# Show the plot
plt.show()
```
